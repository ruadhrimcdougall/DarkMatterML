```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# layer 1 10 filts

loss_1_10 = np.load('Saved Arrays/f2testing/Layer 1/loss_10filters.npy')
acc_1_10 = np.load('Saved Arrays/f2testing/Layer 1/acc_10filters.npy')
wacc_1_10 = np.load('Saved Arrays/f2testing/Layer 1/w_acc_10filters.npy')
f1_1_10 = np.load('Saved Arrays/f2testing/Layer 1/f1_10filters.npy')
f2_1_10 = np.load('Saved Arrays/f2testing/Layer 1/f2_10filters.npy')
f3_1_10 = np.load('Saved Arrays/f2testing/Layer 1/f3_10filters.npy')

conv_sizes_layer1 = np.around(np.logspace(-2, 1.3, 30) * 100)
print(conv_sizes_layer1)
conv_sizes_layer1 = np.delete(conv_sizes_layer1, [1, 3])
```

```python
plt.figure()
plt.semilogx(conv_sizes_layer1 / 100, np.mean(acc_1_10, axis=0), label='accuracy')
plt.semilogx(conv_sizes_layer1 / 100, np.mean(wacc_1_10, axis=0), label='weighted accuracy')
plt.semilogx(conv_sizes_layer1 / 100, np.mean(f1_1_10, axis=0), label='f1 score')
plt.semilogx(conv_sizes_layer1 / 100, np.mean(f2_1_10, axis=0), label='f2 score')
plt.semilogx(conv_sizes_layer1 / 100, np.mean(f3_1_10, axis=0), label='f3 score')
plt.plot(np.zeros(50)+5.38, np.linspace(0.6,0.88), linestyle='dashed', color='grey')
plt.plot(np.zeros(50)+9.09, np.linspace(0.6,0.88), linestyle='dashed', color='grey')
# plt.plot(np.logspace(-2, 1.3), np.zeros(50)+0.78, linestyle='dashed', color='grey')
plt.xlabel(r'Filter Size ($\mu$s)')
plt.ylabel('Metric Score')
```
